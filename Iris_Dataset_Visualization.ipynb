```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv("/content/Iris.csv")
print(iris)
```

```python
iris.drop(columns="Id",axis= 1)
```

plotting 


```python
# sepal length vs sepal width
plot = iris[iris.Species == "Iris-setosa"].plot.scatter(x = "SepalLengthCm",y = "SepalWidthCm",color="r",label = "Iris-setosa")
iris[iris.Species == "Iris-versicolor"].plot.scatter(x = "SepalLengthCm",y = "SepalWidthCm",color = "g",label= "Iris-versicolor",ax = plot)
iris[iris.Species == "Iris-virginica"].plot.scatter(x ="SepalLengthCm",y = "SepalWidthCm",color = "b" , label = "Iris-virginica",ax = plot)
plt.xlabel("SepalLength")
plt.ylabel("SepalWidth")
plt.title("sepal length vs sepal width")
plt.show
```

```python

# Petal Length vs Petal Width

plot = iris[iris.Species=='Iris-setosa'].plot.scatter(x='PetalLengthCm',y='PetalWidthCm',color='r',label='Iris Setosa')
iris[iris.Species=='Iris-versicolor'].plot.scatter(x='PetalLengthCm',y='PetalWidthCm',color='b',label='Iris Versicolor',ax=plot)
iris[iris.Species=='Iris-virginica'].plot.scatter(x='PetalLengthCm',y='PetalWidthCm',color='g',label='Iris Virginica',ax=plot)

plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Petal Length vs Petal Width')
plt.show()
```

```python
setosa = df[df['species']=='setosa']
sns.kdeplot( setosa['sepal_width'], setosa['sepal_length'],
                 cmap="plasma", shade=True, shade_lowest=False)
```

```python
plt.figure(1)
plt.figure(figsize = (15,10))
num_cols = ['sepal_length','sepal_width','petal_length','petal_width']
for col in num_cols:
    plt.subplot(int(str(22)+str((num_cols.index(col)+1))))
    sns.violinplot(x='species', y = col, data = df)
```
